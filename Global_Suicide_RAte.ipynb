```python
import pandas as pd
```

```python
# import data
df=pd.read_csv("/content/30_merged_dataset_v00_final.csv")
df.head()
```

```python
#check how many null
df.isnull().sum()
```

```python
#check dtypes
df.dtypes
```

```python
#check suicide rate
df["suicide rate %"]=df["Suicides number"]/df["Population"]*100
df.head()
```

```python
#check suicide rate by country
suicide_rate_country=pd.DataFrame(df.groupby("Country").mean()["suicide rate"])
suicide_rate_country
```

```python
#check top 5 suicide rate
suicide_rate_country.sort_values(by="suicide rate", ascending=False, inplace=True)
suicide_rate_country[:5]
```

```python
#check bottom 5 suicide rate
suicide_rate_country.sort_values(by="suicide rate", ascending=True, inplace=True)
suicide_rate_country[:5]
```
